## Column Description
It is a description of the columns in the dataset, providing a understanding of each column:


- **User_ID**: Unique identifier for each user in the dataset.

- **Product_ID**: Identifier for each product purchased by the customer.

- **Gender**: Gender of the user, typically represented as "M" for male and "F" for female.

- **Age**: Age group of the user, typically provided in ranges such as "0-17", "18-25", "26-35", etc.

- **Occupation**: Code representing the occupation of the user. Each number corresponds to a specific occupation category.

- **City_Category**: Code ('A', 'B', 'C') indicating the size and type of city the customer resides in like Tier-1, Tier-2, Tier-3 cities.

- **Stay_In_Current_City_Years**: Duration (in years) the customer has lived in the current city, with '4+' representing four or more years.

- **Marital_Status**: Indicates if the customer is married (1) or unmarried (0).

- **Product_Category_1**: Primary category code for the product type (e.g., 1 = Electronics, 3 = Clothing).

- **Product_Category_2**: Secondary, optional sub-category for further product details (e.g., 6 = Accessories for Electronics).

- **Product_Category_3**: Tertiary, optional sub-category adding finer details (e.g., 14 = Premium Accessories).

- **Purchase**: Purchase amount spent by the user on the specific product, represented as a numeric value.

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# warnings
import warnings
warnings.filterwarnings("ignore")

# Load dataset
df = pd.read_csv("https://new-assets.ccbp.in/frontend/content/aiml/classical-ml/classical_ml_sales_dataset.csv")
```

```python
# Preview dataset
print(df.head())
print(df.info())
print(df.describe())
```

Q1: Analyze how the average purchase amount varies across different age groups and genders. Present the results in a tabular format with three columns: Gender, Age Group, Average Purchase. Then, visualize the comparison using a suitable plot (e.g., grouped bar plot).

```python
# Write your code here
avg_purchase_gender_age = df.groupby(["Gender", "Age"])["Purchase"].mean().reset_index()
print(avg_purchase_gender_age)
```

```python
# write your code here for visuals/plots
plt.figure(figsize=(10,6))
ax = sns.barplot(x="Age", y="Purchase", hue="Gender",
                 data=avg_purchase_gender_age, palette="Set2")

plt.title("Average Purchase Amount by Gender & Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Purchase Amount")

# Create a custom legend (this shows which color = which gender)
handles, labels = ax.get_legend_handles_labels()

# Optional: replace labels with more descriptive text
labels = ["Female (0)", "Male (1)"]
ax.legend(handles, labels, title="Gender")

plt.show()
```

Q3: Calculate the total purchase amounts across different city categories. Present the results in a summary table with two columns: `City_Category` and `Purchase`. Finally, create a bar plot to visualize the comparison across cities.

Q4: Determine the age group with the 2nd highest average purchase amount for each gender. Present the results in a summary table with columns: Age Group, Gender, and Average Purchase. Finally, visualize the comparison using a suitable plot (e.g., grouped bar chart).

```python
# Write your code here
age_group = df.groupby(['Age','Gender'])['Purchase'].mean().reset_index()
ranked_age_group = age_group.groupby("Gender").apply(lambda x: x.nlargest(2,"Purchase")).reset_index(drop=True)
print(ranked_age_group)
```

```python
# write your code here for visuals/plots
plt.figure(figsize=(8,5))
ax = sns.barplot(x='Age', y='Purchase', hue='Gender', data=ranked_age_group,
                 palette='Set2', dodge=True)

plt.title('Top 2 Age Groups by Average Purchase (per Gender)')
plt.xlabel('Age Group')
plt.ylabel('Average Purchase Amount')
plt.legend(title='Gender')

plt.show()
```

Q10: Calculate the top 5 occupations by total purchase share, give it in table format with columns: Occupation, Purchase, Percent and visualize their percentage contributions.

```python
# Write your code here
occ_purchase = df.groupby("Occupation")['Purchase'].sum().reset_index()
top5_occ = occ_purchase.sort_values("Purchase",ascending=False).head(5)
top5_occ["Percent"] = (top5_occ["Purchase"]/top5_occ["Purchase"].sum())*100
print(top5_occ)
```

```python
# write your code here for visuals/plots
plt.figure(figsize=(8,6))
sns.barplot(x="Occupation", y="Percent", data=top5_occ)
plt.title("Top 5 Occupations by Purchase %")
plt.show()
```

Q15: Identify which features show the strongest positive and negative correlation with the Purchase amount. Then, generate a heatmap to visualize correlation.

```python
# Write your code here
# Calculate correlation matrix
corr = df.corr(numeric_only=True)

# Identify strongest correlations with Purchase
print(corr["Purchase"].sort_values(ascending=False))
```
